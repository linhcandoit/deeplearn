```python
def add_to_class(Class):
    def wrapper(function):
        setattr(Class, function.__name__, function)
    return wrapper
```

```python
class B(d2l.HyperParameters):
    def __init__(self, a, b, c):
        self.save_hyperparameters(ignore=['c'])
        print(f"a = {a}, b = {b}")
        print(hasattr(self, 'c'))

b = B(1,2,3) 
```
